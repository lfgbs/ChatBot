```python
def process_intent(Intent, input_phrase):
    
    tokens = word_tokenize(input_phrase)
    tag = nltk.pos_tag(tokens)
    spa_text=nlp(input_phrase)
    #print([(X.text, X.label_) for X in spa_text.ents])
    
    if Intent=="BookRestaurant":
        
        date=""
        num_people=""
        restaurant=""
        time=""
    
 
        for entry in spa_text.ents:
            if entry.label_=="DATE":
                date=entry.text
            
            if entry.label_=="CARDINAL":
                num_people=entry.text
                
            if entry.label_=="TIME":
                time=entry.text
                
            if entry.label_=="ORG":
                restaurant=entry.text
                
        if date=="":
            date=input("Please tell me the date for the reservation ")
            
        if time=="":
            time=input("Please tell me the time for the reservation ")
            
        if num_people=="":
            num_people=input("Please tell me for how many people is the reservation ")
            
        if restaurant=="":
            restaurant=input("Please tell me for which restaurant is the reservation ")
            
        print("Making reservation for "+num_people+" for "+date+" at "+time+" at "+restaurant)
        
        
        
        
    elif Intent=="PlayMusic":     
        
        for url in search(input_phrase, tld="co.in", num=1, stop=1, pause=2):
            page=urllib.request.urlopen(url)
            html=BeautifulSoup(page.read(), "html.parser")
            print("Playing "+html.title.string)
            print(url)
            

                
    elif Intent=="SearchCreativeWork":

        for j in search(input_phrase, tld="co.in", num=5, stop=5, pause=2):
            print(j)
            
    elif Intent=="RateBook":
        book=""
        rating_given=""
        max_rating=""
        
        for entry in spa_text.ents:
            if entry.label_=="WORK_OF_ART":
                book=entry.text
            
            if entry.label_=="CARDINAL" and rating_given == "":
                rating_given=entry.text
                
            elif entry.label_=="CARDINAL" and rating_given != "":
                if int(rating_given)>int(entry.text):
                    max_rating=rating_given
                    rating_given=entry.text
                else:
                    max_rating=entry.text
                    
        if book=="": 
            book=input("Please enter the name of the book ")
            
        if rating_given=="":
            rating_given=input("What is the rating being given ")
            
        if max_rating=="":
            max_rating=input("What are you rating out of? ")
            
    
        print("Rated "+book+" "+rating_given+" out of "+max_rating+" stars")
                    
                

                
    elif Intent=="AddToPlaylist":
        music=input_phrase.split("to")[0][3:]
        playlist=input_phrase.split("to")[1]

        print("Adding"+music+"to"+playlist)

    elif Intent=="SearchScreeningEvent":
        
        movie=""
        date=""
        location=""
        
        for entry in spa_text.ents:
            if entry.label_=="DATE":
                date=entry.text
            
            if entry.label_=="GPE":
                location=entry.text
            
        if date=="":
            date=input("You want to check movie sessions for when? ")
            
        if location=="":
            location=input("What location? ")
            
        if movie=="":
            movie=input("What movie? ")
        
            
        query=movie+" "+location+" "+date
                
              
        for j in search(query, tld="co.in", num=3, stop=3, pause=2):         
              print(j)
                
    elif Intent=="GetWeather":
              
        date=""
        location=""
        
        for entry in spa_text.ents:
            if entry.label_=="DATE":
                date=entry.text
            
            if entry.label_=="GPE":
                location=entry.text
                
                
        if date=="":
            date=input("You want to check the weather for when? ")
            
        if location=="":
            location=input("What location? ")
        
            
        print("Checking weather in "+location+" for "+date)
        
    else:
        print("The developers are still working on a response for intent: "+Intent)
                
            
```

```python

def chat(inp, data, last_input):
    
    # load trained model
    model = keras.models.load_model('chat_model')

    # load tokenizer object
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)

    # load label encoder object
    with open('label_encoder.pickle', 'rb') as enc:
        lbl_encoder = pickle.load(enc)

    # parameters
    max_len = 20
    name=""
    intentlist=[]
    
    print(Fore.LIGHTBLUE_EX + "User: " + Style.RESET_ALL, end="")
        
    if inp.lower() == "wrong":
        print("Is it in any of these intents? If not, write a new one")
        if intentlist==[]:
            for i in data['intents']:
                intentlist.append(i['intent'])
                print(i['intent'])
        intents=input()
        if intents in intentlist:
            index=intentlist.index(intents)
            data["intents"][index]["text"].append(last_input)
            print(data["intents"][index]["text"])
            return
        
        print("Please add an answer:")
        newanswer=input()
        temp={"intent":intents,"text":[last_input],"responses":[newanswer]}
        data["intents"].append(temp)
        intentlist.append(intents)
        print("Thanks for helping me")
        return
    result = model(keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([inp]),
                                         truncating='post', maxlen=max_len))
    tag = lbl_encoder.inverse_transform([np.argmax(result)])
    last_tag=tag
    if tag == "GreetingResponse" or tag=="CourtesyGreetingResponse":
        name=inp.split()[-1]
    for i in data['intents']:
        if i['intent'] == tag:
            answer=np.random.choice(i['responses'])
            if name!="":
                answer=answer.replace("<HUMAN>", name)
            print(Fore.GREEN + "ChatBot:" + Style.RESET_ALL , answer)
    last_input=inp
    if tag == "TimeQuery":
        print(ctime())
    # print(Fore.GREEN + "ChatBot:" + Style.RESET_ALL,random.choice(responses))
#print(Fore.YELLOW + "Start messaging with the bot (type quit to stop)!" + Style.RESET_ALL)
```

# Merge

```python
# Function to convert text to
# speech
def SpeakText():
	
	# Exception handling to handle
# exceptions at the runtime
    try:

        # use the microphone as source for input.
        with sr.Microphone() as source2:

            # wait for a second to let the recognizer
            # adjust the energy threshold based on
            # the surrounding noise level
            r.adjust_for_ambient_noise(source2, duration=0.2)

            #listens for the user's input
            audio2 = r.listen(source2)

            # Using google to recognize audio
            MyText = r.recognize_google(audio2)
            MyText = MyText.lower()
            
            return MyText


    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))

    except sr.UnknownValueError:
        print("Sorry, can you repeat?")
	
```
